# Analizando una Imagen

```python
import matplotlib.pyplot as plt 
import numpy as np

# 1. Leer la imagen beagle.jpg
img = plt.imread('beagle.jpg')

# 2. Mostrar las dimensiones de la imagen
print("Dimensiones de la imagen (Alto, Ancho, Canales):", img.shape)

# 3. Mostrar el valor de un píxel específico
pixel_val = img[10,10]
print("Valor del píxel en [10, 10]:", pixel_val)

# 4. Visualizar la imagen
plt.imshow(img)
plt.title('Imagen Original - Beagle')
plt.axis('off')
plt.show()
```

## Ejercicio 1: Canales de Color
Separa la imagen en sus canales Rojo, Verde y Azul (RGB) y muéstralos por separado.

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(img[:, :, 0], cmap='Reds')
axs[0].set_title('Canal Rojo')
axs[0].axis('off')

axs[1].imshow(img[:, :, 1], cmap='Greens')
axs[1].set_title('Canal Verde')
axs[1].axis('off')

axs[2].imshow(img[:, :, 2], cmap='Blues')
axs[2].set_title('Canal Azul')
axs[2].axis('off')

plt.show()
```

## Ejercicio 2: Escala de Grises
Convierte la imagen a escala de grises promediando los canales y muéstrala.

```python
gray_img = np.mean(img, axis=2)
plt.imshow(gray_img, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.axis('off')
plt.show()
```
